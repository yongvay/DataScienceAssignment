## Data Visualization 
Histograms for the numerical columns <br>
Bar charts for the categorical columns

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the numerical columns for plotting
numerical_cols = ['Age', 'PlayTimeHours', 'SessionsPerWeek', 'AvgSessionDurationMinutes', 'PlayerLevel', 'AchievementsUnlocked']

print("Generating Histograms for Numerical Columns...")

# Set up the matplotlib figure and axes for a 2x3 grid
plt.figure(figsize=(16, 10)) # Adjust figure size for better readability

# Loop through each numerical column and create a histogram
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1) # Create a 2x3 grid of subplots (row, col, index)
    sns.histplot(df[col], kde=True, bins=30) # kde=True adds a Kernel Density Estimate line, bins for detail
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', alpha=0.75) # Add grid lines for readability

plt.tight_layout(pad=3.0) # Adjust layout to prevent overlapping titles/labels, add some padding
plt.suptitle('Histograms of Numerical Features', y=1.02, fontsize=16) # Add a super title for the entire figure
plt.show() # Display the plot

print("Histograms generated and displayed.")
```
